```python
from langchain.llms import CTransformers
from langchain import PromptTemplate , LLMChain
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
```

```python
#for csv
#import pandas as pd
#from langchain_experimental.agents import create_csv_agent
```

```python
llm = CTransformers (model="TheBloke/Llama-2-13B-chat-GGML",model_type='llama',config={'max_new_tokens': 600,
                              'temperature': 0.01,
                              'context_length': 5000}, model_file = 'llama-2-13b-chat.ggmlv3.q2_K.bin', callbacks=[ StreamingStdOutCallbackHandler() ])
```

```python
template = """ [INST] <> You are a helpful, respectful and honest assistant. Your answers are always brief. <> {text}[/INST] """

prompt = PromptTemplate(template=template, input_variables=["text"])
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
# User Input
user_input = input("Enter your question: ")

# Pass the input to the LLM chain
response = llm_chain.invoke(user_input)
```
